```python
"""
Football Player Performance Analysis Script
==========================================
A comprehensive script to analyze a single player's performance in one match
using StatsBomb-like event data and lineup information.
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Rectangle
from matplotlib.patches import FancyBboxPatch
import warnings
warnings.filterwarnings('ignore')

# Set style for better-looking plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class PlayerPerformanceAnalyzer:
    """
    A class to analyze and visualize a football player's performance in a single match.
    """
    
    def __init__(self, events_df, lineup_df):
        """
        Initialize the analyzer with event and lineup data.
        
        Parameters:
        events_df (DataFrame): Event-level match data
        lineup_df (DataFrame): Player lineup information
        """
        self.events = events_df.copy()
        self.lineup = lineup_df.copy()
        self.player_events = None
        self.player_info = None
        self.kpis = {}
        
    def filter_player_data(self, match_id, player_name):
        """
        Filter data for a specific player in a specific match.
        
        Parameters:
        match_id: Match identifier
        player_name (str): Name of the player to analyze
        """
        # Filter events for the specific match and player
        self.player_events = self.events[
            (self.events['match_id'] == match_id) & 
            (self.events['player'] == player_name)
        ].copy()
        
        # Get player info from lineup
        self.player_info = self.lineup[
            self.lineup['player_name'] == player_name
        ].iloc[0] if len(self.lineup[self.lineup['player_name'] == player_name]) > 0 else None
        
        if self.player_events.empty:
            print(f"Warning: No events found for player '{player_name}' in match {match_id}")
        
        if self.player_info is None:
            print(f"Warning: No lineup info found for player '{player_name}'")
    
    def calculate_kpis(self):
        """
        Calculate key performance indicators for the player.
        """
        if self.player_events is None or self.player_events.empty:
            return {}
        
        # Minutes played (approximate from last event)
        if not self.player_events.empty:
            minutes_played = self.player_events['minute'].max()
        else:
            minutes_played = 0
        
        # Goals
        goals = len(self.player_events[
            (self.player_events['type'] == 'Shot') & 
            (self.player_events['shot_outcome'] == 'Goal')
        ])
        
        # Assists
        assists = len(self.player_events[
            self.player_events['pass_goal_assist'] == True
        ])
        
        # Shots analysis
        shots = self.player_events[self.player_events['type'] == 'Shot']
        total_shots = len(shots)
        shots_on_target = len(shots[shots['shot_outcome'].isin(['Goal', 'Saved'])])
        shots_off_target = total_shots - shots_on_target
        
        # Expected goals
        xg = shots['shot_statsbomb_xg'].sum() if 'shot_statsbomb_xg' in shots.columns else 0
        
        # Passes analysis
        passes = self.player_events[self.player_events['type'] == 'Pass']
        total_passes = len(passes)
        successful_passes = len(passes[passes['pass_outcome'].isna()])  # Null means successful
        unsuccessful_passes = total_passes - successful_passes
        pass_accuracy = (successful_passes / total_passes * 100) if total_passes > 0 else 0
        
        # Duels
        duels = self.player_events[self.player_events['type'] == 'Duel']
        total_duels = len(duels)
        duels_won = len(duels[duels['duel_outcome'] == 'Won'])
        duels_lost = total_duels - duels_won
        
        # Fouls
        fouls_committed = len(self.player_events[self.player_events['type'] == 'Foul Committed'])
        fouls_won = len(self.player_events[self.player_events['type'] == 'Foul Won'])
        
        # Cards
        yellow_cards = len(self.player_events[self.player_events['foul_committed_card'] == 'Yellow Card'])
        red_cards = len(self.player_events[self.player_events['foul_committed_card'] == 'Red Card'])
        
        self.kpis = {
            'minutes_played': minutes_played,
            'goals': goals,
            'assists': assists,
            'total_shots': total_shots,
            'shots_on_target': shots_on_target,
            'shots_off_target': shots_off_target,
            'xg': round(xg, 2),
            'total_passes': total_passes,
            'successful_passes': successful_passes,
            'unsuccessful_passes': unsuccessful_passes,
            'pass_accuracy': round(pass_accuracy, 1),
            'total_duels': total_duels,
            'duels_won': duels_won,
            'duels_lost': duels_lost,
            'fouls_committed': fouls_committed,
            'fouls_won': fouls_won,
            'yellow_cards': yellow_cards,
            'red_cards': red_cards
        }
        
        return self.kpis
    
    def plot_player_heatmap(self, figsize=(12, 8)):
        """
        Create a heatmap showing player's position on the field.
        """
        if self.player_events is None or self.player_events.empty:
            print("No data available for heatmap")
            return
        
        # Filter events with location data
        events_with_location = self.player_events[
            self.player_events['location'].notna()
        ].copy()
        
        if events_with_location.empty:
            print("No location data available for heatmap")
            return
        
        # Extract x, y coordinates (assuming location is a list [x, y])
        try:
            events_with_location[['x', 'y']] = pd.DataFrame(
                events_with_location['location'].tolist(), 
                index=events_with_location.index
            )
        except:
            print("Could not parse location data for heatmap")
            return
        
        fig, ax = plt.subplots(figsize=figsize)
        
        # Create the pitch
        self._draw_pitch(ax)
        
        # Create heatmap
        heatmap, xedges, yedges = np.histogram2d(
            events_with_location['x'], 
            events_with_location['y'], 
            bins=20, 
            range=[[0, 120], [0, 80]]
        )
        
        # Plot heatmap
        im = ax.imshow(
            heatmap.T, 
            extent=[0, 120, 0, 80], 
            origin='lower', 
            alpha=0.7, 
            cmap='Reds'
        )
        
        plt.colorbar(im, ax=ax, label='Number of Events')
        
        player_name = self.player_info['player_name'] if self.player_info is not None else 'Player'
        ax.set_title(f'{player_name} - Heat Map', fontsize=16, fontweight='bold')
        ax.set_xlabel('Field Length (yards)')
        ax.set_ylabel('Field Width (yards)')
        
        plt.tight_layout()
        plt.show()
    
    def plot_shot_map(self, figsize=(12, 8)):
        """
        Create a shot map showing all shots taken by the player.
        """
        shots = self.player_events[self.player_events['type'] == 'Shot'].copy()
        
        if shots.empty:
            print("No shots to display")
            return
        
        fig, ax = plt.subplots(figsize=figsize)
        self._draw_pitch(ax)
        
        # Extract shot locations
        try:
            shots[['x', 'y']] = pd.DataFrame(
                shots['location'].tolist(), 
                index=shots.index
            )
        except:
            print("Could not parse shot location data")
            return
        
        # Plot shots with different colors for outcomes
        for outcome in shots['shot_outcome'].unique():
            if pd.notna(outcome):
                shot_data = shots[shots['shot_outcome'] == outcome]
                color = 'red' if outcome == 'Goal' else 'orange' if outcome == 'Saved' else 'gray'
                size = 200 if outcome == 'Goal' else 100
                
                ax.scatter(
                    shot_data['x'], shot_data['y'],
                    c=color, s=size, alpha=0.7,
                    label=f'{outcome} ({len(shot_data)})',
                    edgecolors='black', linewidth=1
                )
        
        ax.legend(loc='upper left')
        player_name = self.player_info['player_name'] if self.player_info is not None else 'Player'
        ax.set_title(f'{player_name} - Shot Map', fontsize=16, fontweight='bold')
        ax.set_xlabel('Field Length (yards)')
        ax.set_ylabel('Field Width (yards)')
        
        plt.tight_layout()
        plt.show()
    
    def plot_pass_map(self, figsize=(14, 10), max_passes=50):
        """
        Create a pass map showing pass connections.
        """
        passes = self.player_events[
            (self.player_events['type'] == 'Pass') & 
            (self.player_events['location'].notna())
        ].copy()
        
        if passes.empty:
            print("No pass data to display")
            return
        
        # Limit number of passes for readability
        if len(passes) > max_passes:
            passes = passes.sample(max_passes)
        
        fig, ax = plt.subplots(figsize=figsize)
        self._draw_pitch(ax)
        
        try:
            # Extract pass start locations
            passes[['start_x', 'start_y']] = pd.DataFrame(
                passes['location'].tolist(), 
                index=passes.index
            )
            
            # For end locations, we'll use a simplified approach
            # In real StatsBomb data, you'd use pass end location
            for idx, pass_event in passes.iterrows():
                start_x, start_y = pass_event['start_x'], pass_event['start_y']
                
                # Simplified end location calculation (you'd use actual end location in real data)
                angle = pass_event.get('pass_angle', 0) if pd.notna(pass_event.get('pass_angle')) else np.random.uniform(-np.pi, np.pi)
                length = pass_event.get('pass_length', 20) if pd.notna(pass_event.get('pass_length')) else 20
                
                end_x = start_x + length * np.cos(angle)
                end_y = start_y + length * np.sin(angle)
                
                # Ensure coordinates stay within field bounds
                end_x = np.clip(end_x, 0, 120)
                end_y = np.clip(end_y, 0, 80)
                
                # Color based on pass outcome
                color = 'green' if pd.isna(pass_event['pass_outcome']) else 'red'
                alpha = 0.7 if pd.isna(pass_event['pass_outcome']) else 0.4
                
                ax.annotate('', xy=(end_x, end_y), xytext=(start_x, start_y),
                           arrowprops=dict(arrowstyle='->', color=color, alpha=alpha, lw=1))
        
        except Exception as e:
            print(f"Could not create pass map: {e}")
            return
        
        # Add legend
        from matplotlib.lines import Line2D
        legend_elements = [
            Line2D([0], [0], color='green', lw=2, alpha=0.7, label='Successful Pass'),
            Line2D([0], [0], color='red', lw=2, alpha=0.4, label='Unsuccessful Pass')
        ]
        ax.legend(handles=legend_elements, loc='upper left')
        
        player_name = self.player_info['player_name'] if self.player_info is not None else 'Player'
        ax.set_title(f'{player_name} - Pass Map (Sample of {len(passes)} passes)', fontsize=16, fontweight='bold')
        ax.set_xlabel('Field Length (yards)')
        ax.set_ylabel('Field Width (yards)')
        
        plt.tight_layout()
        plt.show()
    
    def plot_half_comparison(self, figsize=(12, 6)):
        """
        Compare 1st half vs 2nd half performance.
        """
        if self.player_events is None or self.player_events.empty:
            return
        
        # Split data by half
        first_half = self.player_events[self.player_events['period'] == 1]
        second_half = self.player_events[self.player_events['period'] == 2]
        
        # Calculate stats for each half
        stats = {
            'Passes': [
                len(first_half[first_half['type'] == 'Pass']),
                len(second_half[second_half['type'] == 'Pass'])
            ],
            'Shots': [
                len(first_half[first_half['type'] == 'Shot']),
                len(second_half[second_half['type'] == 'Shot'])
            ],
            'Duels': [
                len(first_half[first_half['type'] == 'Duel']),
                len(second_half[second_half['type'] == 'Duel'])
            ]
        }
        
        # Create comparison chart
        fig, ax = plt.subplots(figsize=figsize)
        
        x = np.arange(len(stats))
        width = 0.35
        
        first_half_values = [stats[key][0] for key in stats]
        second_half_values = [stats[key][1] for key in stats]
        
        bars1 = ax.bar(x - width/2, first_half_values, width, label='1st Half', alpha=0.8)
        bars2 = ax.bar(x + width/2, second_half_values, width, label='2nd Half', alpha=0.8)
        
        ax.set_xlabel('Action Type')
        ax.set_ylabel('Count')
        ax.set_title('1st Half vs 2nd Half Performance', fontsize=16, fontweight='bold')
        ax.set_xticks(x)
        ax.set_xticklabels(stats.keys())
        ax.legend()
        
        # Add value labels on bars
        for bar in bars1:
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height,
                   f'{int(height)}', ha='center', va='bottom')
        
        for bar in bars2:
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height,
                   f'{int(height)}', ha='center', va='bottom')
        
        plt.tight_layout()
        plt.show()
    
    def plot_pass_accuracy_pie(self, figsize=(8, 8)):
        """
        Create a pie chart showing pass success vs failure.
        """
        if self.kpis.get('total_passes', 0) == 0:
            print("No pass data available")
            return
        
        fig, ax = plt.subplots(figsize=figsize)
        
        sizes = [self.kpis['successful_passes'], self.kpis['unsuccessful_passes']]
        labels = ['Successful', 'Unsuccessful']
        colors = ['#2ecc71', '#e74c3c']
        explode = (0.05, 0)  # Slightly separate the successful slice
        
        wedges, texts, autotexts = ax.pie(
            sizes, labels=labels, colors=colors, explode=explode,
            autopct='%1.1f%%', startangle=90, shadow=True
        )
        
        # Enhance text
        for autotext in autotexts:
            autotext.set_color('white')
            autotext.set_fontweight('bold')
            autotext.set_fontsize(12)
        
        ax.set_title(
            f'Pass Accuracy: {self.kpis["pass_accuracy"]}%\n({self.kpis["successful_passes"]}/{self.kpis["total_passes"]} passes)',
            fontsize=16, fontweight='bold', pad=20
        )
        
        plt.tight_layout()
        plt.show()
    
    def plot_timeline(self, figsize=(14, 8)):
        """
        Create a timeline of key events during the match.
        """
        if self.player_events is None or self.player_events.empty:
            return
        
        # Define key event types and their colors
        key_events = {
            'Goal': {'events': self.player_events[
                (self.player_events['type'] == 'Shot') & 
                (self.player_events['shot_outcome'] == 'Goal')
            ], 'color': '#2ecc71', 'marker': 'o', 'size': 150},
            'Yellow Card': {'events': self.player_events[
                self.player_events['foul_committed_card'] == 'Yellow Card'
            ], 'color': '#f1c40f', 'marker': 's', 'size': 100},
            'Red Card': {'events': self.player_events[
                self.player_events['foul_committed_card'] == 'Red Card'
            ], 'color': '#e74c3c', 'marker': 's', 'size': 100},
            'Assist': {'events': self.player_events[
                self.player_events['pass_goal_assist'] == True
            ], 'color': '#3498db', 'marker': '^', 'size': 120}
        }
        
        fig, ax = plt.subplots(figsize=figsize)
        
        y_pos = 0
        legend_elements = []
        
        for event_type, event_data in key_events.items():
            events = event_data['events']
            if not events.empty:
                minutes = events['minute'].values
                ax.scatter(
                    minutes, [y_pos] * len(minutes),
                    c=event_data['color'], marker=event_data['marker'],
                    s=event_data['size'], alpha=0.8, edgecolors='black',
                    label=f"{event_type} ({len(events)})"
                )
                
                # Add minute labels
                for minute in minutes:
                    ax.text(minute, y_pos + 0.1, f"{int(minute)}'", 
                           ha='center', va='bottom', fontsize=9)
        
        # Customize the plot
        ax.set_xlim(-5, 95)
        ax.set_ylim(-0.5, 0.5)
        ax.set_xlabel('Match Minute', fontsize=12)
        ax.set_title('Timeline of Key Events', fontsize=16, fontweight='bold')
        ax.set_yticks([])  # Remove y-axis ticks
        ax.grid(True, alpha=0.3)
        ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        
        # Add half-time line
        ax.axvline(x=45, color='gray', linestyle='--', alpha=0.7, label='Half Time')
        
        plt.tight_layout()
        plt.show()
    
    def display_player_info(self):
        """
        Display player information from lineup data.
        """
        if self.player_info is None:
            print("No player information available")
            return
        
        print("=" * 50)
        print("PLAYER INFORMATION")
        print("=" * 50)
        print(f"Name: {self.player_info['player_name']}")
        print(f"Jersey Number: {self.player_info['jersey_number']}")
        print(f"Country: {self.player_info['country']}")
        print(f"Position(s): {self.player_info['positions']}")
        if pd.notna(self.player_info.get('cards')):
            print(f"Cards: {self.player_info['cards']}")
        print("=" * 50)
    
    def display_kpis(self):
        """
        Display calculated KPIs in a formatted way.
        """
        if not self.kpis:
            print("No KPIs calculated")
            return
        
        print("\n" + "=" * 50)
        print("KEY PERFORMANCE INDICATORS")
        print("=" * 50)
        
        # Basic stats
        print(f"Minutes Played: {self.kpis['minutes_played']}")
        print(f"Goals: {self.kpis['goals']}")
        print(f"Assists: {self.kpis['assists']}")
        print(f"Expected Goals (xG): {self.kpis['xg']}")
        print()
        
        # Shooting stats
        print("SHOOTING:")
        print(f"  Total Shots: {self.kpis['total_shots']}")
        print(f"  On Target: {self.kpis['shots_on_target']}")
        print(f"  Off Target: {self.kpis['shots_off_target']}")
        print()
        
        # Passing stats
        print("PASSING:")
        print(f"  Total Passes: {self.kpis['total_passes']}")
        print(f"  Successful: {self.kpis['successful_passes']}")
        print(f"  Unsuccessful: {self.kpis['unsuccessful_passes']}")
        print(f"  Accuracy: {self.kpis['pass_accuracy']}%")
        print()
        
        # Dueling stats
        print("DUELS:")
        print(f"  Total Duels: {self.kpis['total_duels']}")
        print(f"  Won: {self.kpis['duels_won']}")
        print(f"  Lost: {self.kpis['duels_lost']}")
        print()
        
        # Disciplinary stats
        print("DISCIPLINE:")
        print(f"  Fouls Committed: {self.kpis['fouls_committed']}")
        print(f"  Fouls Won: {self.kpis['fouls_won']}")
        print(f"  Yellow Cards: {self.kpis['yellow_cards']}")
        print(f"  Red Cards: {self.kpis['red_cards']}")
        print("=" * 50)
    
    def _draw_pitch(self, ax):
        """
        Helper function to draw a football pitch.
        """
        # Pitch dimensions (in yards)
        pitch_length = 120
        pitch_width = 80
        
        # Set up the figure
        ax.set_xlim(0, pitch_length)
        ax.set_ylim(0, pitch_width)
        ax.set_aspect('equal')
        
        # Draw pitch outline
        pitch = Rectangle((0, 0), pitch_length, pitch_width, 
                         linewidth=2, edgecolor='white', facecolor='green', alpha=0.3)
        ax.add_patch(pitch)
        
        # Draw center line
        ax.plot([pitch_length/2, pitch_length/2], [0, pitch_width], 'white', linewidth=2)
        
        # Draw center circle
        center_circle = plt.Circle((pitch_length/2, pitch_width/2), 10, 
                                  color='white', fill=False, linewidth=2)
        ax.add_patch(center_circle)
        
        # Draw penalty areas
        # Left penalty area
        left_penalty = Rectangle((0, pitch_width/2 - 22), 18, 44,
                               linewidth=2, edgecolor='white', facecolor='none')
        ax.add_patch(left_penalty)
        
        # Right penalty area
        right_penalty = Rectangle((pitch_length-18, pitch_width/2 - 22), 18, 44,
                                linewidth=2, edgecolor='white', facecolor='none')
        ax.add_patch(right_penalty)
        
        # Draw goal areas
        # Left goal area
        left_goal = Rectangle((0, pitch_width/2 - 10), 6, 20,
                            linewidth=2, edgecolor='white', facecolor='none')
        ax.add_patch(left_goal)
        
        # Right goal area
        right_goal = Rectangle((pitch_length-6, pitch_width/2 - 10), 6, 20,
                             linewidth=2, edgecolor='white', facecolor='none')
        ax.add_patch(right_goal)
        
        ax.set_facecolor('darkgreen')
        ax.grid(False)
    
    def generate_full_report(self, match_id, player_name):
        """
        Generate a complete performance report for a player.
        """
        print(f"Generating performance report for {player_name} in match {match_id}")
        print("Please wait...")
        
        # Step 1: Filter data
        self.filter_player_data(match_id, player_name)
        
        # Step 2: Calculate KPIs
        self.calculate_kpis()
        
        # Step 3: Display player info and KPIs
        self.display_player_info()
        self.display_kpis()
        
        # Step 4: Generate all visualizations
        print("\nGenerating visualizations...")
        
        self.plot_player_heatmap()
        self.plot_shot_map()
        self.plot_pass_map()
        self.plot_half_comparison()
        self.plot_pass_accuracy_pie()
        self.plot_timeline()
        
        print("\nPerformance analysis complete!")


# ============================================================================
# EXAMPLE USAGE
# ============================================================================

def main():
    """
    Main function demonstrating how to use the PlayerPerformanceAnalyzer.
    """
    # Step 1: Load your data
    # Replace these with your actual dataframes
    # events = pd.read_csv('your_events_file.csv')
    # lineup = pd.read_csv('your_lineup_file.csv')
    
    # For demonstration, create sample data structure
    print("Creating sample data for demonstration...")
    
    # Sample events data
    events_sample = pd.DataFrame({
        'type': ['Pass', 'Shot', 'Pass', 'Duel', 'Pass'] * 20,
        'player': ['Lionel Messi'] * 100,
        'player_id': [123] * 100,
        'team': ['Barcelona'] * 100,
        'match_id': [1001] * 100,
        'minute': np.random.randint(1, 90, 100),
        'second': np.random.randint(0, 59, 100),
        'shot_outcome': ['Goal', 'Saved', None, None, None] * 20,
        'shot_statsbomb_xg': [0.8, 0.1, None, None, None] * 20,
        'pass_outcome': [None, None, None, None, 'Incomplete'] * 20,
        'pass_length': np.random.randint(5, 40, 100),
        'pass_angle': np.random.uniform(-np.pi, np.pi, 100),
        'pass_goal_assist': [True] + [False] * 99,
        'duel_outcome': [None, None, None, 'Won', None] * 20,
        'foul_committed_card': [None] * 98 + ['Yellow Card', None],
        'location': [[x, y] for x, y in zip(np.random.randint(20, 100, 100), 
                                           np.random.randint(10, 70, 100))],
        'period': [1] * 50 + [2] * 50
    })
    
    # Sample lineup data
    lineup_sample = pd.DataFrame({
        'player_id': [123],
        'player_name': ['Lionel Messi'],
        'jersey_number': [10],
        'country': ['Argentina'],
        'cards': [None],
        'positions': ['Right Wing']
    })
    
    # Step 2: Initialize analyzer
    analyzer = PlayerPerformanceAnalyzer(events_sample, lineup_sample)
    
    # Step 3: Generate complete report
    analyzer.generate_full_report(match_id=1001, player_name='Lionel Messi')


if __name__ == "__main__":
    main()
```
